# Limpieza de datos

```python
import pandas as pd
import json
```

```python
file_data = open('/content/drive/MyDrive/BEDU/csvjson.json','r')
file_data
```

```python
result_data= json.load(file_data)
print(type(result_data))
```

```python
df = pd.DataFrame(result_data)
df.head(10)
```

```python
#Eliminamos columnas que no nos sirven
df_limpio = df.drop(columns=['Calendar Address','Visibility', 'Date Created', 'Last Updated', 'MyStatus', 'All Day Event', 'Recurring Event', 'Recurring Event', 'Created By'])
df_limpio
```

```python
#Mostramos si tenemos espacios vacíos por columna
df_limpio.isna().sum()
```

```python
#Renombramos las columnas
column_names = {
    'Event Title' : 'nombre_cliente',
    'Event Description' : 'descripcion_servicio',
    'Event Start' : 'hora_inicio',
    'Event End' : 'hora_fin',
    'Calculated Duration' : 'duracion_servicio',
    'Event End__1' : 'fecha_servicio'
}
df_limpio = df_limpio.rename(columns=column_names)
df_limpio.head()
```

```python
df_limpio.dtypes
```

```python
df_limpio.loc[df_limpio["descripcion_servicio"]==""]
```

```python
df_limpio.loc[df_limpio["descripcion_servicio"] == "", "descripcion_servicio"] = "Sin comentarios"
df_limpio
```

```python
#Tipo de valor
df_limpio['fecha_servicio'] = pd.to_datetime(df_limpio['fecha_servicio'])
print(df_limpio.dtypes)
```

```python
df_limpio.isna().sum()
```

```python
df_limpio.head()
```

```python
#Estableciendo tipo de datos


df_limpio['hora_inicio'] = pd.to_datetime(df_limpio['hora_inicio'], format='%H:%M:%S', errors='coerce')
df_limpio['hora_fin'] = pd.to_datetime(df_limpio['hora_fin'], format='%H:%M:%S', errors='coerce')
df_limpio['fecha_servicio'] = pd.to_datetime(df_limpio['fecha_servicio'], format='%d/%m/%Y')
print(df_limpio.dtypes)
df_limpio
```

# Funciones estadísticas

```python
#Convirtiendo la columna duracion_servicio a numérico
df_limpio['duracion_servicio'] = pd.to_numeric(df_limpio['duracion_servicio'], errors='coerce')
```

```python
duracion_min = df_limpio['duracion_servicio'].min()
print("La duración mínima del servicio fue de", duracion_min,"hrs")

duracion_max = df_limpio['duracion_servicio'].max()
print("La duración máxima del servicio fue de", duracion_max,"hrs")

duracion_media = df_limpio['duracion_servicio'].mean()
print("La duración media del servicio fue de", duracion_media,"hrs")
```

```python
# Tenemos una discrepancia porque sale -20.0 hrs como valor mínimo
# Checamos primero si en efecto, tenemos ese valor o es error de nuestro código
if -20 in df_limpio['duracion_servicio'].values:
    print("El valor -20 existe en la columna.")
else:
    print("El valor -20 no existe en la columna.")
```

```python
#Ahora checamos cuántos datos tienen el valor -20
cantidad_menos_20 = df_limpio['duracion_servicio'].value_counts()[-20]

print(cantidad_menos_20)
```

```python
#Sustituimos esos valores de -20 por 0
df_limpio['duracion_servicio'] = df_limpio['duracion_servicio'].replace("-20", 0)
```

```python
#Calculamos de nuevos nuestras funciones estadísticas con los valores congruentes actualizados

duracion_min = df_limpio['duracion_servicio'].min()
print("La duración mínima del servicio fue de", duracion_min,"hrs")

duracion_max = df_limpio['duracion_servicio'].max()
print("La duración máxima del servicio fue de", duracion_max,"hrs")

duracion_media = df_limpio['duracion_servicio'].mean()
print("La duración media del servicio fue de", duracion_media,"hrs")
```
